```python
import statistics

# Given data set
data = [100, 200, 300, 500, 900]

# Calculate mean, variance, and standard deviation
mean_value = statistics.mean(data)
variance = statistics.variance(data)  # Sample variance (n-1)
std_deviation = statistics.stdev(data)
                                 
     # Display results
print(f"Mean: {mean_value}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_deviation}")                            
```

```python
import numpy as np

# Given data
x = [100, 200, 300, 500, 900]

# Calculate the mean ()
mu = np.mean(x)

# Calculate variance using the formula Var(X) =  p(x) * (x - )^2
# Since we are dealing with a discrete distribution where p(x) = 1/n for each x in x
n = len(x)
variance = sum([(xi - mu) ** 2 for xi in x]) / n

print(f"variance: {variance}")
```
